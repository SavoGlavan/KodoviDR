```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score,train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
data=pd.read_csv("combined_features.csv")
data
```

```python
pairs=pd.read_csv("PairList.csv")
pairs
```

```python
data["Interacting"]=pairs["Interacting"]
```

```python
features = data.drop(['Unnamed: 0', 'Protein-Pair'], axis=1)
features
```

```python
y=features["Interacting"]
y
```

```python
features=features.drop("Interacting", axis=1)
features
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(features)
scaled
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled, y, test_size=0.3, random_state=42)
X_train
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=5)

# Train the model on the training set
knn_classifier.fit(X_train, y_train)

# Make predictions on the test set
predictions = knn_classifier.predict(X_test)

# Print the classification report
print("Classification Report:")
print(classification_report(y_test, predictions))
```

```python
# Define the parameter grid for the grid search
param_grid = {
    'n_neighbors': [3, 5, 7, 9],  # Adjust as needed
    'weights': ['uniform', 'distance'],  # Weighting scheme
    'p': [1, 2]  # Power parameter for the Minkowski distance (1 for Manhattan, 2 for Euclidean)
}

# Initialize the KNN classifier
knn_classifier = KNeighborsClassifier()

# Perform grid search with 5-fold cross-validation
grid_search = GridSearchCV(knn_classifier, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(scaled, y)

# Print the best parameters and corresponding accuracy
print("Best Parameters:", grid_search.best_params_)
print("Best Accuracy: {:.2f}".format(grid_search.best_score_))

# Train the model on the entire training set using the best parameters
best_knn_classifier = grid_search.best_estimator_
best_knn_classifier.fit(X_train, y_train)

# Make predictions on the test set
predictions = best_knn_classifier.predict(X_test)

# Print the classification report
print("Classification Report:")
print(classification_report(y_test, predictions))
```
